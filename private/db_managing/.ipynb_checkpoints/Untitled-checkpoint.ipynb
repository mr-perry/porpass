```python
import os
from pathlib import Path
from dotenv import load_dotenv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString
import pymysql as sql

import sys
sys.path.append("/Users/mrperry/python/grasp/")
import grasp
# Locate .env two levels up from private/db_managing/
#env_path = Path(__file__).resolve().parents[2] / '.env'
env_path = "/Applications/XAMPP/xamppfiles/htdocs/porpass/.env"
load_dotenv(env_path)

# Then access values like this
db_host = os.getenv('DB_HOST')
db_name = os.getenv('DB_DATABASE')
db_user = os.getenv('DB_USERNAME')
db_pass = os.getenv('DB_PASSWORD')
```

```python


mkFile = "/Volumes/data/NAIF/pds/sln-l-spice-6-v1.0/slnsp_1000/extras/mk/SEL_V02.TM"
grasp.furnish(mkFile)
spice_params = grasp.grab_spice_params('SELENE')
utc = True
obsrvr_str = spice_params['OBSRVRSTR']
fix_ref_str = spice_params['FIXREFSTR']
ab_corr = spice_params['ABCORR']
target_str = spice_params['TARGETSTR']
method = "NEAR POINT/ELLIPSOID"
lrs_archive = Path("/Volumes/data/SELENE/lrs/darts/sln-l-lrs-2-sndr-waveform-high-v1.0/")
lrs_subs = sorted(lrs_archive.glob('200?????'))
data_str = os.path.join("data", "LRS_??_??_*.tbl")
inst_id = int(1)
body_id = int(3)


qry_tmp = '''INSERT INTO observations 
    (instrument_id, body_id, native_id, start_time, end_time, 
     start_lat, start_lon, end_lat, end_lon, 
     observation_geometry, science_file_id)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, ST_GeomFromText(%s, 0), %s)'''

#
# Connect to SQL
#
cnx = sql.connect(user=db_user, password=db_pass, host=db_host, database=db_name)
cursor = cnx.cursor()

dry_run = True
for sub in lrs_subs:
    sci_files = sorted(sub.glob(data_str))
    for sci_file in sci_files:
        try:
            sci = grasp.read(sci_file)
            et = sci['OBSERVATION_TIME']
            vctrs = grasp.compute_state_vectors(et, obsrvr_str, target_str, fix_ref_str,
                                                ab_corr=ab_corr,
                                                intercept_method_str=method, utc=utc)
            vels = grasp.decompose_velocity(vctrs.r_s, vctrs.v_s)
            geom = grasp.compute_geometry(vctrs.r_t, vctrs.r_s, vctrs.r_st)
            coords = [(lon, lat) for lon, lat in zip(geom.longitude, geom.latitude)]
            line = LineString(coords)
            native_id = os.path.basename(sci_file).split(".")[0]
            cursor.execute(qry_tmp, (
                inst_id, body_id, native_id,
                pd.Timestamp(et[0]), pd.Timestamp(et[-1]),
                float(geom.latitude[0]), float(geom.longitude[0]),
                float(geom.latitude[-1]), float(geom.longitude[-1]),
                line.wkt, None
            ))
        except Exception as e:
            print(f"Failed on {sci_file.name}: {e}")
            continue
    break
if dry_run:
    print("Dry run complete -- rolling back")
    cnx.rollback()
else:
    cnx.commit()
    print("Committed successfully")
cnx.close()
```
